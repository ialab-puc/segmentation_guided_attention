```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#constants
DATA_PATH = 'votes_clean.csv'
ATTRIBUTES = ['wealthy',
    'depressing',
    'safety',
    'lively',
    'boring',
    'beautiful'
 ]
```

```python
data = pd.read_csv(DATA_PATH)
```

```python
data
```

# Repetition count

```python
#map data into hash table
_hash = {}
for row in data.iterrows():
    if (row[1]['left_id'],row[1]['right_id'],row[1]['category']) in _hash:
        _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])] +=1
    else:
        _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])] =1
```

```python
reps = np.array(list(_hash.values()))
unique, counts = np.unique(reps, return_counts=True)
dict_counts = dict(zip(unique, counts))
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [9.5, 6]

plt.bar(unique[1:],counts[1:], .5, align='center', alpha=0.5, log=True)
plt.title('Repetition frecuency')
plt.xlabel('N° Repetitions')
plt.ylabel('Frecuency (Log)')
plt.show()
```

```python
attributes_map = { at:{} for at in ATTRIBUTES}
for key,value in _hash.items():
    attributes_map[key[2]][key] = value
    
```

```python
plt.rcParams['figure.figsize'] = [15, 12]
for i,attr in enumerate(ATTRIBUTES):
    reps = np.array(list(attributes_map[attr].values()))
    unique, counts = np.unique(reps, return_counts=True)
    plt.subplot(2,3,i+1)
    plt.bar(unique[1:],counts[1:], .5, align='center', alpha=0.5)
    plt.title(f'Repetition frecuency {attr}')
    plt.xlabel('N° Repetitions')
    plt.ylabel('Frecuency (Log)')
plt.show()
```

# Repetition Uniformity

```python
#map data into hash table
_hash = {}
for row in data.iterrows():
    if (row[1]['left_id'],row[1]['right_id'],row[1]['category']) in _hash:
        _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])][0] += 1
        if row[1]['winner'] == 'left':
            _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])][1] += 1
        elif row[1]['winner'] == 'right':
            _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])][1] += -1
    else:
        if row[1]['winner'] == 'left':
            _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])] =[1,1]
        elif row[1]['winner'] == 'right':
            _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])] =[1,-1]
        else:
            _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])] =[1,0]
```

```python
for key in list(_hash.keys()):
    if _hash[key][0] == 1:
        del _hash[key]
        
for key in _hash.keys():
    _hash[key] = _hash[key][0] == abs(_hash[key][1])
```

```python
reps = np.array(list(_hash.values()))
unique, counts = np.unique(reps, return_counts=True)
counts = counts/len(reps)
plt.bar(np.arange(len(unique)),counts, .5, align='center')
plt.title('Uniform answers')
plt.ylabel('%')
plt.xticks(np.arange(len(unique)), unique)
plt.show()
```

```python
attributes_map = { at:{} for at in ATTRIBUTES}
for key,value in _hash.items():
    attributes_map[key[2]][key] = value
```

```python
plt.rcParams['figure.figsize'] = [15, 12]
for i,attr in enumerate(ATTRIBUTES):
    reps = np.array(list(_hash.values()))
    unique, counts = np.unique(reps, return_counts=True)
    counts = counts/len(reps)
    plt.subplot(2,3,i+1)
    plt.bar(np.arange(len(unique)),counts, .5, align='center')
    plt.title(f'Uniform answers {attr}')
    plt.ylabel('%')
    plt.xticks(np.arange(len(unique)), unique)
plt.show()
```

## Repetition removal

```python
data['delete']=0
# set deletable rows to 1
_hash = {}
for row in data.iterrows():
    if (row[1]['left_id'],row[1]['right_id'],row[1]['category']) in _hash:
        data.loc[row[0], 'delete'] = 1
        if _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])][1] != row[1]['winner']:
            if data.loc[_hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])][0],'delete'] != 1:
                data.loc[_hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])][0],'delete'] = 1
            
    else:
        _hash[(row[1]['left_id'],row[1]['right_id'],row[1]['category'])] = (row[0],row[1]['winner'])
```
